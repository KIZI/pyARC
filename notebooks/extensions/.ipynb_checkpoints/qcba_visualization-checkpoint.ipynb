```python
min_improvement = 0
min_conditional_improvement = -0.01

rule = rule_to_extend
current_best = rule

direct_extensions = qcba_transformation.extender._RuleExtender__get_extensions(rule)
```

```python
direct_extensions
```

```python
current_best.update_properties(quant_dataset)

extension_succesful = False
```

```python
direct_extensions = qcba_transformation.extender._RuleExtender__get_extensions(current_best)

for candidate in direct_extensions:
    candidate.update_properties(quant_dataset)

    delta_confidence = candidate.confidence - current_best.confidence
    delta_support = candidate.support - current_best.support


    if qcba_transformation.extender._RuleExtender__crisp_accept(delta_confidence, delta_support, min_improvement):
        print("crisp accept")
        current_best = candidate
        extension_succesful = True
        break


    if qcba_transformation.extender._RuleExtender__conditional_accept(delta_confidence, min_conditional_improvement):
        enlargement = candidate

        while True:

            enlargement = qcba_transformation.extender.get_beam_extensions(enlargement)

            if not enlargement:
                break

            candidate.update_properties(quant_dataset)
            enlargement.update_properties(quant_dataset)

            delta_confidence = enlargement.confidence - current_best.confidence
            delta_support = enlargement.support - current_best.support

            if qcba_transformation.extender._RuleExtender__crisp_accept(delta_confidence, delta_support, min_improvement):
                current_best = enlargement
                extension_succesful = True

            elif qcba_transformation.extender._RuleExtender__conditional_accept(delta_confidence, min_conditional_improvement):
                continue

            else:
                break


        if extension_succesful == True:
            break


    else:
        # continue to next candidate
        continue


```
